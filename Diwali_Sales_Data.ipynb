# Exploratory Data Analysis

**Gender**


1.   order
2.   Amount



```python
g = sns.countplot(x="Gender", data=df)
for bars in g.containers:
  g.bar_label(bars) # Pass the 'bars' variable (BarContainer) to bar_label
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Count')
plt.show()
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Total Amount by Gender')
plt.show()
```

# Age

```python
age=sns.countplot(x='Age Group',data=df)
for bars in age.containers:
  age.bar_label(bars)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group Count')
plt.show()
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Age Group', y='Amount',data=sales_age)
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.title('Total Amount by Age Group')
plt.show()
```

# State

```python
Sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=Sales_state,x='State',y='Amount')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.title('Total Amount by State')
plt.show()
```

Marital_status

```python
sales_marital=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=sales_marital)
plt.xlabel('Marital Status')
plt.ylabel('Total Amount')
plt.title('Total Amount by Marital Status')
plt.show()
```

```python
marital = sns.countplot(x="Marital_Status", data=df , hue='Gender')
for bars in marital.containers:
  marital.bar_label(bars)
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Marital Status Count')
plt.show()
```

# Occupation

```python
occupation = sns.countplot(x="Occupation", data=df)
for bars in occupation.containers:
  occupation.bar_label(bars)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation Count')
plt.show()
#
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
plt.xlabel('Occupation')
plt.ylabel('Total Amount')
plt.title('Total Amount by Occupation')
plt.show
```
